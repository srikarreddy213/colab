## Average Discount Percentage by Brand

### Subtask:
Calculate the average discount percentage for each brand and display the results.


**Reasoning**:
To calculate the average discount percentage, I first need to determine the discount amount and then the discount percentage for each product. The `discount_amount` is the difference between `actual_price` and `discount_price`, and the `discount_percentage` is the `discount_amount` divided by `actual_price` multiplied by 100.



```python
df['discount_amount'] = df['actual_price'] - df['discount_price']
df['discount_percentage'] = (df['discount_amount'] / df['actual_price']) * 100
print("Calculated discount amount and percentage for each product.")
```

**Reasoning**:
Now that the discount percentage for each product has been calculated, I will group the DataFrame by 'brand' and compute the mean of the 'discount_percentage' for each brand to find the average discount percentage.



```python
average_discount_by_brand = df.groupby('brand')['discount_percentage'].mean().sort_values(ascending=False)
print("Average Discount Percentage by Brand:")
print(average_discount_by_brand)
```

```python
# ------------------------------
# TASK 1  Walmart Sales Analysis
# Data Loading, Merging & Inspection
# ------------------------------

import pandas as pd

# 1. Load CSV files
sales = pd.read_csv("train.csv")        # or sales.csv depending on your file name
features = pd.read_csv("features.csv")
stores = pd.read_csv("stores.csv")

# 2. Convert Date columns to datetime
sales['Date'] = pd.to_datetime(sales['Date'])
features['Date'] = pd.to_datetime(features['Date'])

# 3. Merge datasets (Store + Dept + Date)
df = sales.merge(features, on=["Store", "Date"], how="left")
df = df.merge(stores, on="Store", how="left")

# 4. Display first and last 10 rows
print("\nFIRST 10 ROWS:")
display(df.head(10))

print("\nLAST 10 ROWS:")
display(df.tail(10))

# 5. Basic Info
print("\nSHAPE of Dataset:", df.shape)

print("\nINFO:")
print(df.info())

print("\nDESCRIBE (Numerical Stats):")
display(df.describe())

# 6. Identify Column Types
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
date_cols = df.select_dtypes(include=['datetime64']).columns

print("\nNUMERICAL Columns:")
print(list(numerical_cols))

print("\nCATEGORICAL Columns:")
print(list(categorical_cols))

print("\nDATE Columns:")
print(list(date_cols))

# 7. List unique store types and departments
print("\nUnique Store Types:")
print(df['Type'].unique())

print("\nUnique Departments:")
print(df['Dept'].unique())

# 8. Identify columns needing cleaning
print("\nPossible Columns Requiring Cleaning / Type Conversion:")
dirty_cols = []

for col in df.columns:
    if df[col].isnull().sum() > 0:
        dirty_cols.append(col)

print(dirty_cols)
```

```python
# -----------------------------
# TASK 2  DATA CLEANING
# -----------------------------

import pandas as pd

# 1. Check missing values
print("\nMissing Values in Each Column:\n")
print(df.isnull().sum())

# 2. Fill missing numeric values using **median**
numeric_fill_cols = ["Temperature", "Fuel_Price", "CPI", "Unemployment"]

for col in numeric_fill_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

# 3. Fill missing Markdown columns using **mean**
markdown_cols = [col for col in df.columns if "MarkDown" in col]

for col in markdown_cols:
    df[col] = df[col].fillna(df[col].mean())

# 4. Convert Date column to datetime format (if not already)
if df['Date'].dtype != "datetime64[ns]":
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# 5. Drop duplicate rows
df = df.drop_duplicates()

# 6. Reset index
df = df.reset_index(drop=True)

print("\nData Cleaning Completed.")
print("New Shape:", df.shape)
```

```python
# -----------------------------------------
# TASK 3  Outlier Detection & Treatment
# -----------------------------------------

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

outlier_cols = ["Weekly_Sales", "Temperature", "Fuel_Price", "CPI"]


# 1 BOX PLOTS FOR VISUAL OUTLIER DETECTION
plt.figure(figsize=(15, 8))
for i, col in enumerate(outlier_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df[col], color="skyblue")
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()


# 2 IQR METHOD FOR OUTLIER DETECTION
def detect_outliers_IQR(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound


# Store results
iqr_results = {}

for col in outlier_cols:
    outliers, lower, upper = detect_outliers_IQR(df, col)
    iqr_results[col] = (len(outliers), lower, upper)
    print(f"{col}: Outliers = {len(outliers)}, Lower = {lower:.2f}, Upper = {upper:.2f}")


# 3 CHECK FOR SALES SPIKES DURING HOLIDAY WEEKS
print("\nHoliday Flag Value Counts:")
print(df["IsHoliday_x"].value_counts())

holiday_sales = df[df["IsHoliday_x"] == True]["Weekly_Sales"]
nonholiday_sales = df[df["IsHoliday_x"] == False]["Weekly_Sales"]

print("\nAvg Weekly Sales During Holidays:", holiday_sales.mean())
print("Avg Weekly Sales During Non-Holidays:", nonholiday_sales.mean())

plt.figure(figsize=(10,5))
sns.boxplot(x=df["IsHoliday_x"], y=df["Weekly_Sales"])
plt.title("Sales Distribution During Holiday vs Non-Holiday Weeks")
plt.show()


# 4 OUTLIER TREATMENT  CAP EXTREME VALUES (WINSORIZATION)
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = np.where(df[column] < lower_bound, lower_bound,
                 np.where(df[column] > upper_bound, upper_bound,
                 df[column]))
    return df


# Capping only numeric continuous variables
for col in ["Temperature", "Fuel_Price", "CPI"]:
    df = cap_outliers(df, col)

print("\nOutlier capping completed.")


# 5 DO NOT REMOVE WEEKLY_SALES OUTLIERS (IMPORTANT BUSINESS OUTLIERS)
print("\nNOTE: Weekly_Sales outliers are KEPT because they may be holiday spikes or promotions.")
```
